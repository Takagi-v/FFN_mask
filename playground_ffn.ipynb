```python
import numpy as np
import matplotlib.pyplot as plt
import torch
from safetensors import safe_open
import os
import seaborn as sns

def load_mask_tensor(file_path):
    """
    加载掩码张量文件，支持.safetensors和.pt格式
    """
    if file_path.endswith('.safetensors'):
        with safe_open(file_path, framework="pt") as f:
            # 大多数情况下，张量名为"tensor"
            tensor = f.get_tensor("tensor")
    elif file_path.endswith('.pt') or file_path.endswith('.bin'):
        tensor = torch.load(file_path, map_location='cpu')
    else:
        raise ValueError(f"不支持的文件格式: {file_path}")
    
    return tensor

def reshape_mask_tensor(tensor, n_layers, intermediate_size):
    """
    将掩码张量重塑为适合热力图显示的形式
    
    Args:
        tensor: 原始掩码张量
        n_layers: 模型层数
        intermediate_size: 每层的intermediate_size
        
    Returns:
        重塑后的tensor，形状为[n_layers, intermediate_size]
    """
    # 确保张量是一维的，并且长度符合预期
    if tensor.dim() > 1:
        # 如果是批次张量，取第一个样本
        tensor = tensor[0].flatten()
    else:
        tensor = tensor.flatten()
    
    expected_length = n_layers * intermediate_size
    if tensor.shape[0] != expected_length:
        print(f"警告: 张量大小 {tensor.shape[0]} 与预期大小 {expected_length} 不符")
        # 如果不符，尝试调整层数
        adjusted_n_layers = tensor.shape[0] // intermediate_size
        print(f"调整为: {adjusted_n_layers} 层 x {intermediate_size} 神经元")
        n_layers = adjusted_n_layers
    
    # 将一维张量重新塑造为 [n_layers, intermediate_size]
    reshaped_tensor = tensor.reshape(n_layers, intermediate_size)
    
    # 应用sigmoid激活，使值在0-1之间
    sigmoid_values = torch.sigmoid(reshaped_tensor)
    
    return sigmoid_values

def plot_heatmap(data, idx=0):
    """
    绘制掩码热力图，类似于您提供的格式
    
    参数:
    - data: sigmoid后的掩码数据，形状为[n_layers, intermediate_size]
    - idx: 训练百分比或步骤索引，用于标题
    """
    data_np = data.cpu().numpy()
    n_layers, n_intermediate = data_np.shape
    
    # 创建包含两个子图的figure, 左右显示
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(20, 10), gridspec_kw={'width_ratios': [1, 1]})
    
    # 翻转数据，使得层0在底部，层n_layers-1在顶部
    heatmap_data = data_np[::-1, :]
    
    # 绘制左侧的热力图 - 原始sigmoid值
    cax1 = ax1.imshow(heatmap_data, cmap='viridis', aspect='auto', vmin=0., vmax=1.)
    fig.colorbar(cax1, ax=ax1, fraction=0.046, pad=0.04)
    ax1.set_title("Sigmoid激活值", fontsize=16)
    
    # 设置左侧热力图的刻度标签
    ax1.set_yticks([])
    ax1.set_xticks(np.arange(0, n_intermediate, n_intermediate//10))
    
    # 添加左侧图表的左轴标签 - 显示层号
    ax_left1 = ax1.twinx()
    ax_left1.yaxis.tick_left()
    ax_left1.set_ylim(0, n_layers)
    ax_left1.set_yticks(np.arange(n_layers) + 0.5)
    ax_left1.set_yticklabels(np.arange(n_layers)[::-1], fontsize=10)
    ax_left1.set_ylabel("层索引", fontsize=14)
    ax_left1.yaxis.set_label_position("left")
    
    # 绘制右侧的热力图 - 二值化后的掩码 (sigmoid值 >= 0.5)
    cax2 = ax2.imshow((heatmap_data >= 0.5).astype(float), cmap='viridis', aspect='auto', vmin=0., vmax=1.)
    fig.colorbar(cax2, ax=ax2, fraction=0.046, pad=0.04)
    ax2.set_title("二值化掩码 (Sigmoid值 >= 0.5)", fontsize=16)
    
    # 设置右侧热力图的刻度标签
    ax2.set_yticks([])
    ax2.set_xticks(np.arange(0, n_intermediate, n_intermediate//10))
    
    # 添加右侧图表的左轴标签 - 显示层号
    ax_left2 = ax2.twinx()
    ax_left2.yaxis.tick_left()
    ax_left2.set_ylim(0, n_layers)
    ax_left2.set_yticks(np.arange(n_layers) + 0.5)
    ax_left2.set_yticklabels(np.arange(n_layers)[::-1], fontsize=10)
    ax_left2.set_ylabel("层索引", fontsize=14)
    ax_left2.yaxis.set_label_position("left")
    
    # 设置整体标题
    fig.suptitle(f"Qwen2 掩码层热力图", fontsize=20)
    
    # 调整布局，为suptitle留出空间
    plt.tight_layout(rect=[0, 0, 1, 0.96])
    
    # 在notebook中显示
    plt.show()
    
    return fig

def plot_layer_activation_rates(data):
    """
    绘制每层激活率的条形图
    
    参数:
    - data: sigmoid后的掩码数据
    """
    data_np = data.cpu().numpy()
    n_layers = data_np.shape[0]
    
    # 计算每层的激活率（大于0.5的比例）
    activation_rates = (data_np >= 0.5).mean(axis=1) * 100  # 转换为百分比
    
    # 绘制条形图
    plt.figure(figsize=(15, 8))
    bars = plt.bar(np.arange(n_layers), activation_rates, color='royalblue')
    
    # 添加数值标签
    for i, v in enumerate(activation_rates):
        plt.text(i, v + 1, f"{v:.1f}%", ha='center', fontsize=9)
    
    # 添加标题和标签
    plt.title("每层掩码激活率", fontsize=18)
    plt.xlabel("层索引", fontsize=14)
    plt.ylabel("激活率 (%)", fontsize=14)
    plt.xticks(np.arange(0, n_layers, max(1, n_layers//20)))
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    
    # 显示图表
    plt.tight_layout()
    plt.show()
    
    return plt.gcf()

# 使用示例 (这部分可以根据您的需要进行修改)
# 请替换为您的实际文件路径和参数
# 示例用法
mask_file = "./output/qwen2-3b/xnli_local_wo/en_zh/checkpoint-12380/model.safetensors"
n_layers = 36  # Qwen2.5-3B-Instruct模型的层数
intermediate_size = 11008  # 每层的intermediate_size

# 加载掩码张量
tensor = load_mask_tensor(mask_file)
print(f"加载的张量形状: {tensor.shape}")

# 重塑张量
sigmoid_values = reshape_mask_tensor(tensor, n_layers, intermediate_size)
print(f"重塑后的张量形状: {sigmoid_values.shape}")

mask_file = "./output/qwen2-3b/xnli_local/zh_en/checkpoint-12380/model.safetensors"
heatmap_fig = plot_heatmap(sigmoid_values)

# 绘制激活率图表
activation_fig = plot_layer_activation_rates(sigmoid_values)

# 输出每层的统计数据
stats = {
    "层索引": list(range(n_layers)),
    "激活率(%)": ((sigmoid_values >= 0.5).float().mean(dim=1) * 100).cpu().numpy().tolist(),
    "平均值": sigmoid_values.mean(dim=1).cpu().numpy().tolist(),
    "最大值": sigmoid_values.max(dim=1)[0].cpu().numpy().tolist(),
    "最小值": sigmoid_values.min(dim=1)[0].cpu().numpy().tolist()
}

# 输出统计数据
import pandas as pd
pd.DataFrame(stats).set_index("层索引")
```

```python
import numpy as np
import matplotlib.pyplot as plt
import torch
from safetensors import safe_open
import os
import seaborn as sns
import pandas as pd

def load_mask_tensor(file_path):
    """
    加载掩码张量文件，支持.safetensors和.pt格式
    """
    if file_path.endswith('.safetensors'):
        with safe_open(file_path, framework="pt") as f:
            # 大多数情况下，张量名为"tensor"
            tensor = f.get_tensor("tensor")
    elif file_path.endswith('.pt') or file_path.endswith('.bin'):
        tensor = torch.load(file_path, map_location='cpu')
    else:
        raise ValueError(f"不支持的文件格式: {file_path}")
    
    return tensor

def reshape_mask_tensor(tensor, n_layers, intermediate_size):
    """
    将掩码张量重塑为适合热力图显示的形式
    
    Args:
        tensor: 原始掩码张量
        n_layers: 模型层数
        intermediate_size: 每层的intermediate_size
        
    Returns:
        重塑后的tensor，形状为[n_layers, intermediate_size]
    """
    # 确保张量是一维的，并且长度符合预期
    if tensor.dim() > 1:
        # 如果是批次张量，取第一个样本
        tensor = tensor[0].flatten()
    else:
        tensor = tensor.flatten()
    
    expected_length = n_layers * intermediate_size
    if tensor.shape[0] != expected_length:
        print(f"警告: 张量大小 {tensor.shape[0]} 与预期大小 {expected_length} 不符")
        # 如果不符，尝试调整层数
        adjusted_n_layers = tensor.shape[0] // intermediate_size
        print(f"调整为: {adjusted_n_layers} 层 x {intermediate_size} 神经元")
        n_layers = adjusted_n_layers
    
    # 将一维张量重新塑造为 [n_layers, intermediate_size]
    reshaped_tensor = tensor.reshape(n_layers, intermediate_size)
    
    # 应用sigmoid激活，使值在0-1之间
    sigmoid_values = torch.sigmoid(reshaped_tensor)
    
    return sigmoid_values

def plot_layer_heatmaps(data, num_layers_to_show=6):
    """
    分别绘制几个代表性层的热力图
    
    参数:
    - data: sigmoid后的掩码数据
    - num_layers_to_show: 要显示的层数
    """
    data_np = data.cpu().numpy()
    n_layers, n_intermediate = data_np.shape
    
    # 选择几个代表性的层进行展示
    layers_to_show = np.linspace(0, n_layers-1, num_layers_to_show, dtype=int)
    
    # 创建一个大的图表
    fig, axes = plt.subplots(num_layers_to_show, 2, figsize=(20, 4*num_layers_to_show))
    
    for i, layer_idx in enumerate(layers_to_show):
        # 获取该层的数据
        layer_data = data_np[layer_idx]
        
        # 绘制sigmoid值
        im1 = axes[i, 0].imshow(layer_data.reshape(1, -1), cmap='viridis', aspect='auto', vmin=0, vmax=1)
        axes[i, 0].set_title(f"第 {layer_idx} 层 Sigmoid激活值")
        axes[i, 0].set_yticks([])
        
        # 绘制二值化掩码
        im2 = axes[i, 1].imshow((layer_data >= 0.5).reshape(1, -1), cmap='viridis', aspect='auto', vmin=0, vmax=1)
        axes[i, 1].set_title(f"第 {layer_idx} 层 二值化掩码")
        axes[i, 1].set_yticks([])
    
    # 添加颜色条
    fig.colorbar(im1, ax=axes[:, 0], fraction=0.015, pad=0.04)
    fig.colorbar(im2, ax=axes[:, 1], fraction=0.015, pad=0.04)
    
    plt.tight_layout()
    plt.show()
    
    return fig

def plot_layer_section(data, layer_idx, start_idx=0, length=1000):
    """
    绘制某一层的局部区域热力图
    
    参数:
    - data: sigmoid后的掩码数据
    - layer_idx: 要展示的层索引
    - start_idx: 起始神经元索引
    - length: 要展示的神经元数量
    """
    data_np = data.cpu().numpy()
    
    # 获取该层数据的子集
    layer_data = data_np[layer_idx]
    section_data = layer_data[start_idx:start_idx+length]
    
    fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(20, 6))
    
    # 绘制sigmoid值
    im1 = ax1.imshow(section_data.reshape(1, -1), cmap='viridis', aspect='auto', vmin=0, vmax=1)
    ax1.set_title(f"第 {layer_idx} 层 Sigmoid激活值 (神经元 {start_idx}-{start_idx+length-1})")
    ax1.set_yticks([])
    
    # 绘制二值化掩码
    im2 = ax2.imshow((section_data >= 0.5).reshape(1, -1), cmap='viridis', aspect='auto', vmin=0, vmax=1)
    ax2.set_title(f"第 {layer_idx} 层 二值化掩码 (神经元 {start_idx}-{start_idx+length-1})")
    ax2.set_yticks([])
    
    # 添加颜色条
    fig.colorbar(im1, ax=ax1, fraction=0.015, pad=0.04)
    fig.colorbar(im2, ax=ax2, fraction=0.015, pad=0.04)
    
    plt.tight_layout()
    plt.show()
    
    return fig

def compare_layers(data):
    """
    对比各层的激活率，并生成统计信息
    
    参数:
    - data: sigmoid后的掩码数据
    """
    data_np = data.cpu().numpy()
    n_layers = data_np.shape[0]
    
    # 计算每层的激活率（大于0.5的比例）
    activation_rates = (data_np >= 0.5).mean(axis=1) * 100  # 转换为百分比
    
    # 计算层统计数据
    layer_stats = pd.DataFrame({
        "层索引": list(range(n_layers)),
        "激活率(%)": activation_rates,
        "平均值": data_np.mean(axis=1),
        "最大值": data_np.max(axis=1),
        "最小值": data_np.min(axis=1),
        "标准差": data_np.std(axis=1)
    }).set_index("层索引")
    
    # 绘制各层激活率对比
    plt.figure(figsize=(15, 8))
    bars = plt.bar(np.arange(n_layers), activation_rates, color='royalblue')
    
    # 添加数值标签
    for i, v in enumerate(activation_rates):
        plt.text(i, v + 1, f"{v:.1f}%", ha='center', fontsize=9)
    
    # 添加标题和标签
    plt.title("各层掩码激活率对比", fontsize=18)
    plt.xlabel("层索引", fontsize=14)
    plt.ylabel("激活率 (%)", fontsize=14)
    plt.xticks(np.arange(0, n_layers, max(1, n_layers//20)))
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    
    plt.tight_layout()
    plt.show()
    
    return layer_stats

def analyze_mask(mask_file, n_layers, intermediate_size):
    """
    综合分析掩码文件
    
    参数:
    - mask_file: 掩码文件路径
    - n_layers: 模型层数
    - intermediate_size: 每层的intermediate_size
    """
    print(f"正在分析掩码文件: {mask_file}")
    
    # 加载掩码张量
    tensor = load_mask_tensor(mask_file)
    print(f"加载的张量形状: {tensor.shape}")
    print(f"张量总元素数: {tensor.numel()}")
    
    # 验证参数是否匹配
    total_size = n_layers * intermediate_size
    if total_size != tensor.numel():
        print(f"警告: 指定参数 ({n_layers} 层 x {intermediate_size} 神经元 = {total_size}) 与张量大小 ({tensor.numel()}) 不匹配")
        
        # 尝试自动推断参数
        if tensor.numel() % n_layers == 0:
            intermediate_size = tensor.numel() // n_layers
            print(f"调整为: {n_layers} 层 x {intermediate_size} 神经元")
        elif tensor.numel() % intermediate_size == 0:
            n_layers = tensor.numel() // intermediate_size
            print(f"调整为: {n_layers} 层 x {intermediate_size} 神经元")
        else:
            # 尝试其他常见的中间层大小
            for test_size in [8192, 11008, 12288, 16384]:
                if tensor.numel() % test_size == 0:
                    n_layers = tensor.numel() // test_size
                    intermediate_size = test_size
                    print(f"调整为: {n_layers} 层 x {intermediate_size} 神经元")
                    break
    
    # 重塑张量
    sigmoid_values = reshape_mask_tensor(tensor, n_layers, intermediate_size)
    print(f"重塑后的张量形状: {sigmoid_values.shape}")
    
    # 计算各层激活率
    print("\n### 各层激活率统计 ###")
    layer_stats = compare_layers(sigmoid_values)
    display(layer_stats)
    
    # 绘制代表性层的热力图
    print("\n### 绘制代表性层的热力图 ###")
    plot_layer_heatmaps(sigmoid_values, num_layers_to_show=6)
    
    # 展示指定层的局部区域
    plot_first_n_neurons = min(1000, intermediate_size)
    print(f"\n### 展示第0层的前{plot_first_n_neurons}个神经元 ###")
    plot_layer_section(sigmoid_values, layer_idx=0, start_idx=0, length=plot_first_n_neurons)
    
    print(f"\n### 展示第{n_layers//2}层的前{plot_first_n_neurons}个神经元 ###")
    plot_layer_section(sigmoid_values, layer_idx=n_layers//2, start_idx=0, length=plot_first_n_neurons)
    
    print(f"\n### 展示第{n_layers-1}层的前{plot_first_n_neurons}个神经元 ###")
    plot_layer_section(sigmoid_values, layer_idx=n_layers-1, start_idx=0, length=plot_first_n_neurons)
    
    return sigmoid_values, layer_stats

# 使用示例
mask_file = "./output/qwen2-3b/xnli_local_test/ar_en/checkpoint-6190/model.safetensors"
n_layers = 36
intermediate_size = 11008  # 修正的intermediate_size

# 执行分析
sigmoid_values, layer_stats = analyze_mask(mask_file, n_layers, intermediate_size)

# 交互式探索 - 可以根据需要取消注释以下代码
"""
# 查看特定层的区域 (修改参数以查看不同层或不同区域)
layer_to_view = 12  # 例如查看第12层
start_neuron = 3000  # 起始神经元索引
num_neurons = 1000   # 查看的神经元数量

plot_layer_section(sigmoid_values, layer_idx=layer_to_view, 
                   start_idx=start_neuron, length=num_neurons)

# 查看激活率最高和最低的层
highest_activation_layer = layer_stats["激活率(%)"].idxmax()
lowest_activation_layer = layer_stats["激活率(%)"].idxmin()

print(f"激活率最高的层: {highest_activation_layer}, 激活率: {layer_stats.loc[highest_activation_layer, '激活率(%)']:.2f}%")
print(f"激活率最低的层: {lowest_activation_layer}, 激活率: {layer_stats.loc[lowest_activation_layer, '激活率(%)']:.2f}%")

# 查看这两层的图形
plot_layer_section(sigmoid_values, layer_idx=highest_activation_layer, start_idx=0, length=1000)
plot_layer_section(sigmoid_values, layer_idx=lowest_activation_layer, start_idx=0, length=1000)
"""
```

```python
import torch
from transformers import AutoModelForCausalLM, AutoTokenizer
from safetensors import safe_open
import pandas as pd

# 设置设备
device = "cuda" if torch.cuda.is_available() else "cpu"
print(f"使用设备: {device}")

def load_mask_tensor(file_path):
    """加载掩码张量文件"""
    if file_path.endswith('.safetensors'):
        with safe_open(file_path, framework="pt") as f:
            tensor = f.get_tensor("tensor")
    elif file_path.endswith('.pt') or file_path.endswith('.bin'):
        tensor = torch.load(file_path, map_location='cpu')
    else:
        raise ValueError(f"不支持的文件格式: {file_path}")
    return tensor

def reshape_mask_tensor(tensor, n_layers, intermediate_size):
    """将掩码张量重塑为适合应用的形式"""
    # 确保张量是一维的
    if tensor.dim() > 1:
        tensor = tensor[0].flatten()
    else:
        tensor = tensor.flatten()
    
    # 重塑为 [n_layers, intermediate_size]
    reshaped_tensor = tensor.reshape(n_layers, intermediate_size)
    
    # 应用sigmoid激活
    sigmoid_values = torch.sigmoid(reshaped_tensor)
    
    return sigmoid_values

def test_translation(model, tokenizer, mask_tensor, texts, n_layers, intermediate_size, src_lang, tgt_lang):
    """执行翻译测试，比较有无掩码的结果"""
    # 重塑并二值化掩码
    mask = reshape_mask_tensor(mask_tensor, n_layers, intermediate_size)
    binary_mask = (mask >= 0.5).float().to(device)
    
    results = []
    
    for text in texts:
        # 准备两种输入：有指令和无指令
        direct_input = text
        instruction_input = f"Translate from {src_lang} to {tgt_lang}: {text}"
        
        inputs_direct = tokenizer(direct_input, return_tensors="pt").to(device)
        inputs_with_instruction = tokenizer(instruction_input, return_tensors="pt").to(device)
        
        result_entry = {"源文本": text}
        
        # 1. 无指令无掩码
        with torch.no_grad():
            outputs_direct_no_mask = model.generate(
                **inputs_direct,
                max_length=256,
                do_sample=False,
                num_beams=5
            )
        result_entry["无指令无掩码"] = tokenizer.decode(outputs_direct_no_mask[0], skip_special_tokens=True)
        
        # 2. 无指令有掩码
        with torch.no_grad():
            outputs_direct_with_mask = model.generate(
                **inputs_direct,
                max_length=256,
                weight_tensor=binary_mask,
                do_sample=False,
                num_beams=5
            )
        result_entry["无指令有掩码"] = tokenizer.decode(outputs_direct_with_mask[0], skip_special_tokens=True)
        
        # 3. 有指令无掩码
        with torch.no_grad():
            outputs_instruction_no_mask = model.generate(
                **inputs_with_instruction,
                max_length=256,
                do_sample=False,
                num_beams=5
            )
        result_entry["有指令无掩码"] = tokenizer.decode(outputs_instruction_no_mask[0], skip_special_tokens=True)
        
        # 4. 有指令有掩码
        with torch.no_grad():
            outputs_instruction_with_mask = model.generate(
                **inputs_with_instruction,
                max_length=256,
                weight_tensor=binary_mask,
                do_sample=False,
                num_beams=5
            )
        result_entry["有指令有掩码"] = tokenizer.decode(outputs_instruction_with_mask[0], skip_special_tokens=True)
        
        results.append(result_entry)
    
    return pd.DataFrame(results)

# 主要执行代码
base_model_path = "/public/share/model/Qwen2.5-3B-Instruct"
mask_file = "./output/qwen2-3b/xnli_local_test/ar_en/checkpoint-6190/model.safetensors"
n_layers = 36
intermediate_size = 11008

# 加载掩码
print("加载掩码...")
mask_tensor = load_mask_tensor(mask_file)
print(f"掩码张量形状: {mask_tensor.shape}, 元素数: {mask_tensor.numel()}")

# 加载模型和分词器
print("加载模型和分词器...")
tokenizer = AutoTokenizer.from_pretrained(base_model_path)
model = AutoModelForCausalLM.from_pretrained(
    base_model_path,
    torch_dtype=torch.float16 if device == "cuda" else torch.float32,
    device_map="auto" if device == "cuda" else None
)
model.eval()

# 测试文本
ar_test_texts = [
    "مرحبا كيف حالك؟",  # "Hello, how are you?"
    "أنا سعيد بلقائك.",  # "I am happy to meet you."
]

# 执行翻译测试
results = test_translation(
    model, tokenizer, mask_tensor, 
    ar_test_texts, n_layers, intermediate_size,
    "Arabic", "English"
)

# 显示结果
print("\n翻译测试结果:")
display(results)
```
